```python
df = df.dropna()
```

```python
df = df.drop_duplicates()
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
df['start_date'] = pd.to_datetime(df['start_date'], errors='coerce')
df['end_date'] = pd.to_datetime(df['end_date'], errors='coerce')
```

```python
df['plan_tier'] = (df['plan_tier'].str.strip().str.title())
df['billing_frequency'] = (df['billing_frequency'].str.strip().str.title())
```

```python
df['seats'] = pd.to_numeric(df['seats'], errors='coerce')
```

```python
df['mrr_amount'] = pd.to_numeric(df['mrr_amount'], errors='coerce')
df['arr_amount'] = pd.to_numeric(df['arr_amount'], errors='coerce')
```

```python
df.loc[df['seats'] <=0, 'seats'] = np.nan
```

```python
df.loc[df['mrr_amount'] < 0, 'mrr_amount'] = np.nan
df.loc[df['arr_amount'] < 0, 'arr_amount'] = np.nan
```

```python
df['subscription_duration_days'] = (df['end_date'] - df['start_date']).dt.days
```

```python
df['mrr_per_seat'] = df['mrr_amount'] / df['seats']
```

```python
df = df.sort_values(['account_id', 'start_date'], ascending=[True, True])
```

```python
df['total_subscriptions'] = df.groupby('account_id')['subscription_id'].transform('count')
```

```python
df = df.reset_index(drop=True)
```

```python
df.to_csv("cleaned_ravenstack_subscriptions.csv",index=False)
```
